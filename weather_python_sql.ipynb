```python
from sqlalchemy import text

create_geo_weather_sql = """
DROP TABLE IF EXISTS geo_weather;

CREATE TABLE geo_weather AS
SELECT
    country,
    location_name,
    latitude,
    longitude,
    timezone,
    last_updated,
    temperature_celsius,
    temperature_fahrenheit,
    condition_text,
    wind_mph,
    wind_kph,
    wind_degree,
    wind_direction,
    pressure_mb,
    pressure_in,
    precip_mm,
    precip_in,
    humidity,
    cloud,
    feels_like_celsius,
    feels_like_fahrenheit,
    visibility_km,
    visibility_miles,
    uv_index,
    gust_mph,
    gust_kph
FROM weather_data;
"""

with engine.connect() as conn:
    conn.execute(text(create_geo_weather_sql))
    conn.commit()

print("geo_weather 建立成功")
```

```python
create_geo_air_quality_sql = """
DROP TABLE IF EXISTS geo_air_quality;

CREATE TABLE geo_air_quality AS
SELECT
    country,
    location_name,
    latitude,
    longitude,
    timezone,
    last_updated,
    "air_quality_Carbon_Monoxide",
    "air_quality_Ozone",
    "air_quality_Nitrogen_dioxide",
    "air_quality_Sulphur_dioxide",
    "air_quality_PM2.5",
    "air_quality_PM10",
    "air_quality_us-epa-index",
    "air_quality_gb-defra-index"
FROM weather_data;
"""

with engine.connect() as conn:
    conn.execute(text(create_geo_air_quality_sql))
    conn.commit()

print("geo_air_quality 建立成功")
```

```python
create_geo_astronomy_sql = """
DROP TABLE IF EXISTS geo_astronomy;

CREATE TABLE geo_astronomy AS
SELECT
    country,
    location_name,
    latitude,
    longitude,
    timezone,
    last_updated,
    sunrise,
    sunset,
    moonrise,
    moonset,
    moon_phase,
    moon_illumination
FROM weather_data;
"""

with engine.connect() as conn:
    conn.execute(text(create_geo_astronomy_sql))
    conn.commit()

print("geo_astronomy 建立成功")
```

```python
#篩選北半球的國家 依照經緯度>0去篩選 並顯示資料筆數及欄位數
geo_weather_north = df[df['latitude'] > 0].copy()
len(geo_weather_north)
geo_weather_north.shape
```

```python
#依照經緯度分類氣候帶
#| 氣候帶               | 緯度範圍        |
#| ----------------- | ----------- |
#| 熱帶 (Tropical)     | 0° – 23.5°  |
#| 亞熱帶 (Subtropical) | 23.5° – 35° |
#| 溫帶 (Temperate)    | 35° – 60°   |
#| 寒帶 (Polar)        | 60° 以上      |

def climate_zone(lat):
    if lat <= 23.5:
        return "Tropical"
    elif lat <= 35:
        return "Subtropical"
    elif lat <= 60:
        return "Temperate"
    else:
        return "Polar"

geo_weather_north["climate_zone"] = geo_weather_north["latitude"].apply(climate_zone)
```

```python
#計算各氣候帶有多少筆資料
geo_weather_north["climate_zone"].value_counts()
```

```python
#建立日期與月份欄位。
df['last_updated'] = pd.to_datetime(df['last_updated'])

df['date'] = df['last_updated'].dt.date
df['month'] = df['last_updated'].dt.to_period('M')
```

```python
def classify_climate_zone(lat):
    lat = abs(lat)
    if lat < 23.5:
        return 'Tropical'
    elif lat < 35:
        return 'Subtropical'
    elif lat < 66.5:
        return 'Temperate'
    else:
        return 'Polar'

df['climate_zone'] = df['latitude'].apply(classify_climate_zone)
```

```python
print(df[['latitude', 'climate_zone', 'date']].head())
print(df['climate_zone'].value_counts())
```

```python
#第一層分析：氣候帶 (Climate Zone)
#每日平均 / 最高 / 最低溫
level1_daily = df.groupby(['climate_zone', 'date']).agg(
    avg_temp=('temperature_celsius', 'mean'),
    max_temp=('temperature_celsius', 'max'),
    min_temp=('temperature_celsius', 'min')
).reset_index()

print(level1_daily.head())
```
