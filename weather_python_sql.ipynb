```python
#將天氣資料寫入sql並建立weather_data table
import pandas as pd

file_path = r"C:\Users\B460M\.cache\kagglehub\datasets\nelgiriyewithana\global-weather-repository\versions\893\GlobalWeatherRepository.csv"

df = pd.read_csv(file_path)

print("CSV 讀取成功")
print("資料大小：", df.shape)
print("欄位名稱：")
print(df.columns)

df.to_sql(
    "weather_data",
    engine,
    if_exists="replace",
    index=False
)

print("weather_data 已成功寫入 PostgreSQL")
```

```python
#確認有需要的weather_data資料表
tables_df = pd.read_sql("""
SELECT table_schema, table_name
FROM information_schema.tables
WHERE table_schema = 'public'
ORDER BY table_name;
""", engine)

print(tables_df)
```

```python
#確認資料有正確匯入
check_df = pd.read_sql("SELECT * FROM weather_data LIMIT 5;", engine)
print(check_df)
```

```python
#確認被匯入的資料欄位是何種型態
columns_df = pd.read_sql("""
SELECT column_name, data_type
FROM information_schema.columns
WHERE table_name = 'weather_data'
ORDER BY ordinal_position;
""", engine)

print(columns_df)
```

```python
#查詢資料總筆數
count_df = pd.read_sql("SELECT COUNT(*) AS total_rows FROM weather_data;", engine)
print(count_df)
```
